```python
import scipy.misc
import numpy as np
import matplotlib.pyplot as plt
face = scipy.misc.face()
```

# Resize the image

```python
print("original size:", face.shape)
yfactor = 5
xfactor = 2
resized = face.repeat(yfactor, axis=0).repeat(xfactor, axis=1)
print("resized:",resized.shape)
```

```python
#plot original
plt.subplot(211)
plt.title('Original face')
plt.axis('off')
plt.imshow(face)

#plot resized
plt.subplot(212)
plt.title('Resized face')
plt.imshow(resized)

#show the result
plt.show()
```
